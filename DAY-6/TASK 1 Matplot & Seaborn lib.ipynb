Matplot lib: Mat plot define the data in graphical.
It is used for draw the chart of given character or co-ordinates

syntax: import matplot

```python
#plot
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([0,2])
y=np.array([0,5])
plt.plot(x,y)
plt.show
```

```python
#wave
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,3,5])
y=np.array([1,4,5])
plt.plot(x,y)
plt.show
plt.plot(x,y,marker="o")
```

```python
#Marker colour
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 4])
y = np.array([5, 8])
plt.plot(x, y, 'o:r')  
plt.show()
```

```python
#Lines Style
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,4,6,8,10])
y = np.array([1,3,5,7,9])
plt.plot(x, y,ls=':')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('title')
plt.grid()
plt.show()
```

```python
#Scatter
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,3,5,8,10,12,16,20])
y=np.array([1,4,5,11,13,18,22,23])
plt.scatter(x,y)
```

```python
#Bar
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A', 'B', 'C', 'D'])
y=np.array([3,5,2,6])
plt.bar(x,y)
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(145,2,270)
plt.hist(x)
plt.show()
```

```python
#pie
import matplotlib.pyplot as plt
import numpy as np
x = np.array([25, 45, 15, 55])
labels = ['A', 'B', 'C', 'D']
plt.pie(x, labels=labels)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('fmri')
sns.lineplot(data=data, x='timepoint', y='signal', hue='region')
plt.title('Line Plot of Signal over Time')
plt.show()
```

```python
# Create a bar plot
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('tips')
sns.barplot(data=data, x='day', y='total_bill')
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
# Scatter
data = sns.load_dataset('tips')
sns.scatterplot(data=data, x='total_bill', y='tip', hue='day')
plt.title('Scatter Plot')
plt.show()
```

```python
#Count
import seaborn as sns
data = sns.load_dataset('tips')
sns.countplot(x='day', data=data)
plt.title('Count Plot')
plt.show()
```

```python
#regression line
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#violinplot
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
import seaborn as sns
data = sns.load_dataset('tips')
sns.pointplot(x='day', y='total_bill', data=data)
plt.title('Point Plot')
plt.show()
```
